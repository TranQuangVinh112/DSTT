```python
#Bài 1:
from scipy import linalg
import numpy as np
a = np.array([1, 2, 3])
b = np.array([[1, 2, 3], [4, 5, 6]])
c = np.array([[(0.5, 1.5, 10), (3, 2, 1)], [(6, 5, 4), (7, 8, 9)]])

A = np.matrix(np.random.random((2, 2)))
B = np.asmatrix(b)
C = np.asmatrix(np.random.random((10, 5)))
D = np.asmatrix([[4, 3], [2, 6]])

F = np.eye(3, k=1)
print(F)

F = np.eye(3, k=0)
print(F)

F = np.eye(3, k=-1)
print(F)

print(np.linalg.matrix_rank(C))

print(A.I)
print(linalg.inv(A))

print(linalg.det(A))
print(A.T)
print(A.H)

b_vec = np.array([1, 2])
print(linalg.solve(D, b_vec))
E = np.asmatrix(a).T
print(E)
print(linalg.lstsq(F, E))
print(np.add(A, D))
print(np.subtract(A, D))
print(np.divide(A, D))

print(A @ D)
print(np.multiply(D, A))
print(np.dot(A, D))
print(np.vdot(A, D))
print(linalg.expm(A))
print(linalg.logm(A))
```

```python
#Bài 2:
import numpy as np
#Ma trận phân kỳ:
A = np.array([[0, 1], [1, 0]])
print(A)
print('---')

temp = A.dot(A)
print(temp)
print('---')

k = 6
for i in range(k - 1):
    temp = temp.dot(A)
    print(temp)
    print('---')

B = np.array([[0, -1], [-1, 0]])
print(B)
print('---')

temp = B.dot(B)
print(temp)
print('---')

k = 5
for i in range(k - 1):
    temp = temp.dot(B)
    print(temp)
    print('---')
```

```python
#Ma trận hội tụ - Convergent matrix:
import numpy as np

C = np.array([ [1, 0, 0], [0, 0.5, 1], [0, 0, 0.5] ])
print(C)
print('---')

temp = C.dot(C)
print(temp)
print('---')

k = 1000
for i in range(k-1):
    temp = temp.dot(C)
print(temp)
print('---')

for i in range(k-1):
    temp = temp.dot(C)
print(temp)
print('---')
```

```python
#Ma trận Markov
import numpy as np

M = np.array([[0.8, 0.3], [0.2, 0.7]])
MM = M.dot(M)  # M^2
print(MM)
print('---')

MM = MM.dot(M)  # M^3
print(MM)
print('---')

for i in range(100):
    MM = MM.dot(M)
print(MM)
print('---')

M = np.array([[0.5, 0.5], [0.5, 0.5]])
MM = M.dot(M)  # M^2
print(MM)
print('---')

MM = MM.dot(M)  # M^3
print(MM)
print('---')

for i in range(100):
    MM = MM.dot(M)
print(MM)
print('---')
```

```python
#Câu 1:
import numpy as np

ma_tran = np.array([[2, 1, 1],
                    [1, 3, 1],
                    [1, 1, 2]])

dinh_thuc = np.linalg.det(ma_tran)
print("Định thức của ma trận:", round(dinh_thuc, 2))
assert dinh_thuc != 0, "Ma trận không khả nghịch!"

chuoi_goc = input("Nhập họ tên hoặc MSSV: ")
mang_ma_ascii = [ord(ky_tu) for ky_tu in chuoi_goc]

while len(mang_ma_ascii) % 3 != 0:
    mang_ma_ascii.append(0)

danh_sach_vector = [mang_ma_ascii[i:i+3] for i in range(0, len(mang_ma_ascii), 3)]
danh_sach_ma_hoa = [np.dot(ma_tran, np.array(vec)) for vec in danh_sach_vector]

print("\nDữ liệu đã mã hóa:")
for v in danh_sach_ma_hoa:
    print(v)

ma_tran_nghich_dao = np.linalg.inv(ma_tran)
danh_sach_giai_ma = [np.dot(ma_tran_nghich_dao, v).round().astype(int) for v in danh_sach_ma_hoa]

chuoi_giai_ma = ''.join([chr(gia_tri) for v in danh_sach_giai_ma for gia_tri in v if gia_tri != 0])
print("\nGiải mã được:")
print(chuoi_giai_ma)
```

```python
#Câu 2:
#a:
import numpy as np
def tinh_phan_so(danh_sach_c):
    M = np.array([[danh_sach_c[0], 1], [1, 0]], dtype=int)
    for c in danh_sach_c[1:]:
        M = np.dot(M, np.array([[c, 1], [1, 0]], dtype=int))
    return M[0][0], M[1][0]
ds = [3, 1, 2]
p, q = tinh_phan_so(ds)
print(f"Phân số: {p}/{q}")
```
